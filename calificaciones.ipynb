```python
import pandas as pd
```

Notas del taller en orden de MARES

```python
tl=pd.read_excel('https://github.com/ComputationalMethods/Evaluacion_2021-1/raw/main/NotasTaller_2021.xlsx')
```

JSON de calificaciones

```python
df=pd.read_json('https://raw.githubusercontent.com/ComputationalMethods/Evaluacion_2021-1/main/calificaciones.json',orient='records')
```

`merge` las dos bases de datos usando la columna `'student_id'` y sólo para los datos de la base de datos izquierda, `tl`→ `how='left'`

```python
df=tl.merge(df,on='student_id',how='left')
```

```python
for k in [c for c in df.columns if c.find('Parcial')>-1 or c.find('Trabajo_Final')>-1]:
    df[f"{k.split('_')[0][0]}{k.split('_')[-1][0]}"]=df[k].apply(lambda d: d.get('value'))
```

```python
df['T']=df['Tareas'].apply(lambda l:sum([d.get('value') for d in l if d.get('Tarea')])/5)
```

```python
df['TT']=df.apply(lambda row: round((row['Taller']+row['T'])/2,2),axis='columns')
```

```python
df[['student_id','P1','P2','P3','TT']].fillna(0)
```
