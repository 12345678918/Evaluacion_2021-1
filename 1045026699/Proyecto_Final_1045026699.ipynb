# Fundameto teórico

Se considera un proyectil  que sale desde el piso en campo plano, que es disparado con una velocidad inicial Vi y regresa al mismo nivel, La trayectoria del proyectil siempre es una parábola.

Por consiguiente se tienen unos parámetros que son necesarios para describir el movimiento de proyectil y son los siguientes; la altura máxima (ecuación 1), el alcance (ecuación 2) y el tiempo máximo (ecuación 3):

1. 
$$
\ gh = g0(\frac{re}{re +h})^2
$$

2. 
$$
R= \frac{(Vi^2- \sin 2\theta )}{g}
$$

3. 
$$
T = \frac{Vi- \sin \theta}{g}
$$

La trayectoria de un proyectil es parabólica y la ecuación de la trayectoria del proyectil que describe este movimiento (no se consideran el roce con el viento ni otras fuerzas adicionales asociadas con el movimiento de la tierra)

las eacuaciones que nos ayudan a describir esto son las siguientes
4. 
$$
\ y = (\tan\alpha)x - \frac{g}{2(Vi\cos\alpha)^2}x^2
$$
5. 
$$
\ y = ax - bx^2
$$

Donde tenemos los puntos notables de la trayectoria parabólica descrita por los proyectiles:

1. La altura máxima
2. el tiempo de vuelo sabiendo que se es descrito por la ecuacion 
$$
\ Tv = 2Tmax
$$
3. El alcance máximo horizontal

el objetivo será por medio de codicos obtener estos datos notables en una trayectoria parabólica y graficar dependiendo las condiciones iniciales


```python
# Librerías para las operaciones matemáticas
from matplotlib.pylab import *
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
# Introducción de datos
while True:
  try:
    grado = float(input("Introducir el ángulo deseado en grados = "))

  except ValueError:
    print("Cantidad de angulo ingresada erronea")

    continue
  break

while True:
  try:

    veli = float(input("introduzca tambien la velocidad inicial deseada en (m/s)"))
  
  except ValueError:
    print("Cantidad de velocidad ingresada erronea")

    continue
  break

print()

#conversion de grados a radianes

angulo = ((grado*np.pi)/180)
print ("θ=",angulo)

g = 9.81 # la gravedad en m/s**2

print()

a = tan(angulo)
b = ((g)/((2*veli**2)*cos(angulo)**2))

#Distancia X y altura maxima

ymax = (veli**2)*(np.sin(angulo)*sin(angulo))/(2*g)
xmax = (veli**2)*(np.sin(2*angulo))/(g)

#Tiempo Maximo y tiempo de vuelo

tmax=(veli*sin(angulo))/(g)
tv=2*(tmax)

#Ahora ponemos la salida de los datos calculados para que se puedan ver

print (str("La altura máxima  alcanzada por el proyectil es: Ymax")+" = "+str(ymax)+" m")
print (str("El alcance máximo horizontal  del proyectil es: Xmax")+" = "+str(xmax)+" m")
print()

print ("El tiempo máximo t1max (s) que alcanza el proyectil para el ángulo θ es: t1max =",format(tmax,".2f"))
print ("El tiempo de vuelo t1v(s) que alcanza el proyectil para el angulo θ es: t1v =",format(tv,".2f"))
print()

#Ecuación de la trayectoria
def f(x):
  return (a*x-b*x**2)

x = np.linspace(0,xmax,500)
```

```python
#Realizamos la figura

plt.figure("Física del movimiento",figsize=(10,8),facecolor="y",edgecolor="c")
plt.axes()

title("LANZAMIENTO DE PROYECTILES", 
fontsize=15,color="blue",verticalalignment="baseline",horizontalalignment = "center")  
plt.suptitle("CINEMATICA",fontsize=20,color="red")

xlabel("xmax",fontsize=20,color="blue")                                      
ylabel("ymax",fontsize=20,color="blue")

plt.grid(True)                                                              
plt.grid(color = '0.5', linestyle = '--', linewidth = 1)

plt.text(((np.argmax(f(x)))/2),np.max(f(x))+1,"vi:",fontsize=10)
plt.text(((np.argmax(f(x)))/2)+11,np.max(f(x))+1,(str(veli)+"m/s"),fontsize=10)

plt.text(3,1,grado,fontsize=10)
plt.plot(x, f(x), "red", linewidth = 2, label = (str(grado)+"º")) 
plt.plot(x,f(x),color="red")

plt.legend(loc = 4,fontsize=10)

valores = [[format(np.max(xmax),".2f"),format(np.min(ymax),".2f")]]
etiquetas_col = ["xmax (m)", "ymax (m)"]
plt.table(cellText=valores, colLabels = etiquetas_col, colWidths = [0.15]*len(f(x)),loc='upper right')


plt.show()

```
