```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
import random 
```

```python
df = pd.read_json("https://github.com/restrepo/anomaly/raw/main/solutions.json.gz")
```

```python
# Punto 1
df["smax"] = df["solution"].apply(lambda x: max(map(abs,x)))
# Punto 2
df["sum_s"] = df["solution"].apply(lambda y: sum(y))
df["sum_s^3"] = df["sum_s"].apply(lambda z:z**3)
```

```python
# Punto 3
random.seed(1001050237)
aleatory = random.randint(0,390073)
coeficients = df.iloc[390073].solution
poly = np.poly1d(coeficients)
roots = poly.roots
realroots = [np.real(x) for x in roots if not np.iscomplex(x)]
deriv_poly = poly.deriv()
```

```python
plt.figure(figsize=(8,8))
x = np.linspace(-1.2,1.1,1000)
y = poly(x)
deriv_y = deriv_poly(x)
plt.plot(x,y, label = "Polinomio")
plt.plot(x,deriv_y, label = "Derivada")
plt.plot(realroots,poly(realroots),"ro", label = "Raices del polinomio")
plt.ylim(min(y),max(y))
plt.xlabel("X")
plt.ylabel("Y")
plt.title(r"$16x^7-17x^6-17x^5+19x^4+20x^3-22x^2-22x+23$")
plt.grid()
plt.legend()
```

La ecuacion es:
\begin{equation}
\frac{\sin x}{x}= -p
\end{equation}

```python
#Parte III
p = random.uniform(0,1)
```

```python
# Parte IV
#Puede no tener raices
equation_roots = []
def ecuation(x,p):
  return (np.sin(x)/x)+p
x_ = np.linspace(-10,10,1000)
for i in x_:
  try:
    new_root = optimize.newton(ecuation,i,args=[p]).round(10)
    if new_root not in equation_roots:
      equation_roots.append(new_root)
  except RuntimeError:
    pass
```

```python

plt.figure(figsize=(8,8))
plt.plot(x_,ecuation(x_,p),label = r"$\frac{\cos x}{x}$")
plt.plot(equation_roots,ecuation(equation_roots,p),"ro",label = "roots")
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(fontsize = 20)
```
