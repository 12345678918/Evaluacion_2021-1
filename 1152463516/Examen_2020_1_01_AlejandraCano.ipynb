```python
# A
import pandas as pd

dfSol = pd.read_json('solutions.json.gz', compression='gzip')  # read and decompress
```

```python
# 1
import numpy as np

dfSol['smax'] = dfSol.solution.apply(lambda n: max(np.abs(n)))  # maximum val of solution col

# 2

dfSol['smax_s'] = dfSol.solution.apply(lambda n: sum(n))  # sum of solution col
dfSol['smax_s3'] = dfSol.solution.apply(lambda n: sum(n)**3)  # sum raised to the 3th
dfSol
```

```python
# 3
import random 
import numpy as np

random.seed(1152463516)

coef = dfSol.solution[random.randint(0, 390073)]
f = np.poly1d(coef)
roots = np.real(np.roots(coef))
roots
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)  # MY interval
df = np.polyder(f)  # derivative of polynomial

plt.style.use('bmh')
plt.plot(x, f(x), color='#006374')
plt.plot(x, df(x), color='#B8860B')
plt.scatter(roots, f(roots), color='#A60628', marker='D')  # graph of roots

plt.xlabel('x')
plt.ylabel('f (x)')
#plt.title('ID Polynomial')  # because its based on my ID
plt.legend(['Polynomial', 'Derivative', 'Roots'])

plt.show()
```
