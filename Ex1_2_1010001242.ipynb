```python
import pandas as pd
import numpy as np  #IMPORTACION DE LIBRERIAS.

df=pd.read_json("https://github.com/restrepo/anomaly/raw/main/solutions.json.gz") #Importamos Toda la base de datos.
df
```

```python
#VALOR MAXIMO
def MAX(x):
    m=sorted( np.abs(x) ) #Definimos Funcion Para El Valor Absoluto.
    return m[-1]
#COLUMNAS
df["N_Max"]=df["solution"].apply(MAX)          #Columna Con El Valor Maximo En Valor Absoluto.
df["Sum_1"]=df["solution"].apply(sum)          #Columna con la suma de Solution.
df["Sum_3"]=(df["solution"].apply(sum))**3     #Columna Con La Suma, Al Cubo, del label Solution.

#GRaficacion nuevo Dataframe.
df
```
